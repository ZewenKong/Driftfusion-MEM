```python
import schemdraw
import schemdraw.elements as elm
```

```python
with schemdraw.Drawing() as d:
    elm.style(elm.STYLE_IEC) # plot style
    d.config(unit=1.5)
    elm.Ground().down()
    elm.Source().right().reverse().flip()
    elm.Line().length(0.75)
    elm.Schottky().flip().label("+$V_{sc}$ (I)", loc='bottom')
    # elm.Line()
    elm.ResistorVar().label("$V_{RPP}$ (I, $R_{RPP}$)")
    elm.Line().length(0.45)
    elm.Schottky().label("-$V_{sc}$ (I)", loc='bottom').reverse()
    elm.Line().length(0.75)
    elm.Ground().up()

    d.save('circuit_plot.png', transparent=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.linspace(0, 2 * np.pi, 1000)
y = np.sin(x)

# Convert cm to inches
cm_to_inch = 1 / 2.5
width_in = 1 * cm_to_inch
height_in = 0.8 * cm_to_inch  # You can adjust the height as needed

# Plot
fig, ax = plt.subplots(figsize=(width_in, height_in), dpi=250)
ax.plot(x, y, color='black', linewidth=2)

# Remove axes and padding
ax.axis('off')
fig.subplots_adjust(left=0, right=1, top=1, bottom=0)

# Save as transparent PNG
fig.savefig("sine_wave.png", transparent=True, bbox_inches='tight', pad_inches=0)
```
