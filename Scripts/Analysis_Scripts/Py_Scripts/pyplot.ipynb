```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img1 = mpimg.imread('C2C.png')
img2 = mpimg.imread('D2D.png')

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

fig.subplots_adjust(left=0, right=1, top=1, bottom=0, wspace=0, hspace=0)

axes[0].imshow(img1)
axes[0].axis('off')

axes[1].imshow(img2)
axes[1].axis('off')

plt.savefig('variability.png', bbox_inches='tight', pad_inches=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.gridspec as gridspec

img1 = mpimg.imread('./pyplot/struc.png')
img2 = mpimg.imread('./pyplot/band.png')
img3 = mpimg.imread('./pyplot/circuit.png')

fig = plt.figure(figsize=(10, 6))
gs = gridspec.GridSpec(1, 3, width_ratios=[1, 2, 1.75])  # fixed: (1, 3)

ax1 = plt.subplot(gs[0])
ax2 = plt.subplot(gs[1])
ax3 = plt.subplot(gs[2])

fig.subplots_adjust(left=0, right=1, top=1, bottom=0, wspace=0, hspace=0)

# Show image 1
ax1.imshow(img1)
ax1.axis('off')

# Show image 2
ax2.imshow(img2)
ax2.axis('off')

# Show image 3
ax3.imshow(img3)
ax3.axis('off')

plt.savefig('device.png', bbox_inches='tight', pad_inches=0)
plt.show()

```
