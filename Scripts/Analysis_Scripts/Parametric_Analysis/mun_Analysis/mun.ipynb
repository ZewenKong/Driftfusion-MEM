```python
import matplotlib.pyplot as plt
from PIL import Image

# Paths to your images
paths = [
    "./mun_cat.png",
    "./mun_thickness.png",
    "./mun_lat.png",
    "./contour.png"
]

# Load images
images = [Image.open(p) for p in paths]

# Create 2x2 grid
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

for ax, img in zip(axs.ravel(), images):
    ax.imshow(img)
    ax.axis('off')  # hide axis for cleaner display

plt.tight_layout()
plt.savefig('mun.png', bbox_inches='tight', pad_inches=0)
plt.show()
```
